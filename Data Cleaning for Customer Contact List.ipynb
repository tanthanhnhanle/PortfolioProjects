# Data Cleaning for Customer Contact List

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\nhanl\Data Analytics\Python\Pandas\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
#df['Last_Name'] = df['Last_Name'].str.lstrip('...')
#df['Last_Name'] = df['Last_Name'].str.lstrip('/')
#df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df['Last_Name'] = df['Last_Name'].str.strip('._/')
df

# lstrip to strip on the left
# rstrip to strip on the right
# strip to strip both side (put all the chars you wanted to strip inside
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x)) # turning the values in the columns into strings
# This replace anything other than a-z, 0-9 into empty
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex = True)  # set regex = True or it won't work
# This adds - in btwn the numbers
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) # adding the '-' in btwn the #
# These replace nan-- and Na-- with empty
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','', regex = True) 
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','', regex = True)
df
```

```python
# This splits the street address into 3 dif col
df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',', expand = True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
#df = df.replace('N/a','')

df = df.fillna('') # fills in all the N/a, None with empty spaces
df
```

```python

# Looping through the values and dropped the customers that did not want to get contacted
# or did not have phone number
for i in df.index:
    if df.loc[i,'Do_Not_Contact'] == 'Y':
        df.drop(i, inplace =True)
    elif df.loc[i,'Do_Not_Contact'] == 'Yes':
        df.drop(i, inplace =True)
    elif df.loc[i,'Do_Not_Contact'] == '':
        df.drop(i, inplace =True)
    elif df.loc[i, 'Phone_Number'] == '':
        df.drop(i, inplace =True)
df

# Another way to drop N/A values
#df = df.dropna(subset = 'Phone_Number', inplace = True)
```

```python
# Reset the index count
df = df.reset_index(drop = True)
df
```
